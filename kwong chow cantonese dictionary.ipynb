```python
import pandas as pd
import numpy as np
import re
```

```python
PATH = r'/home/lun/csrp/dictionaries/'
dict_file = r'cidian_zhyue-kfcd/cidian_zhyue-ft-kfcd-gjyb-2013122.txt'

df = pd.read_csv(PATH + dict_file, index_col=False, header=None, skiprows=range(9), encoding='utf-8')
```

```python
df
```

```python
def removeUnwantedText(text):
    pattern = re.compile('\t.+', flags=re.UNICODE)
    return pattern.sub("", text, re.UNICODE)
```

```python
df[0] = df[0].apply(removeUnwantedText)
```

```python
df[0][df[0].str.contains("[A-Za-z]").values == True].values
```

```python
# create two null columns
df[1] = df[2] = np.nan
```

```python
df.to_csv(PATH + "kwongchow_cantonese_dict.txt", sep=' ', header=False, index=False)
```
